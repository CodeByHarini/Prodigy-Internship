```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
data = pd.read_csv("Task 1 Dataset - World Population.csv", skiprows=4)  
```

```python
# Show first few rows
print("Dataset preview:")
display(data.head())
```

```python
# Example 1: Histogram of Population for a given year
year = "2020"
population_2020 = data[year].dropna()

plt.figure(figsize=(8,5))
plt.hist(population_2020 / 1e6, bins=20, edgecolor='black')  # population in millions
plt.xlabel("Population (Millions)")
plt.ylabel("Number of Countries")
plt.title(f"Distribution of Population across Countries ({year})")
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Example 2: Bar chart for Top 10 most populated countries in 2020
top_countries = data[["Country Name", year]].dropna().sort_values(by=year, ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top_countries["Country Name"], top_countries[year] / 1e9, color="skyblue")
plt.xlabel("Country")
plt.ylabel("Population (Billions)")
plt.title(f"Top 10 Most Populated Countries ({year})")
plt.xticks(rotation=45)
plt.show()
```
